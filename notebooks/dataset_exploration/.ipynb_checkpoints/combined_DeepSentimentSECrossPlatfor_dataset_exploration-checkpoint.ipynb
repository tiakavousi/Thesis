## 2. Loading the Dataset from Excel

In this cell, we load the dataset from the specified Excel file path. Update the path if necessary.

```python
import pandas as pd

file_path = "/Users/tayebekavousi/Desktop/github_sa/datasets/not used/multiple/download/combined_DeepSentimentSECrossPlatform.csv"

try:
    df = pd.read_csv(file_path, encoding="utf-8", engine="python")
    print("Dataset loaded successfully!")
    print("Dataset shape:", df.shape)
except Exception as e:
    print("Error loading dataset:", e)
```

## 3. Display the First 30 Rows

Quickly inspect the dataset by printing the first 10 rows.

```python
print("Head of the dataset (30 rows):")
display(df.head(30))
```

## 4. Sampling 3 Rows from Each Class

We'll sample 3 rows per class (-1= negative, 0 = neutral, 1 = positive) to see a quick preview of each category.

```python
if 'sentiment' in df.columns:
    try:
        # Group by 'oracle' label and sample 3 rows per class
        sample_df = df.groupby('sentiment').apply(lambda x: x.sample(n=3, random_state=42)).reset_index(drop=True)
        print("Sample (3 rows per class):")
        display(sample_df)
    except ValueError as e:
        print("Sampling error:", e)
else:
    print("Column 'sentiment' not found in the dataset.")
```

## 5. Distribution Analysis of Classes

We visualize the distribution of `sentiment` classes using a bar chart.

```python
import matplotlib.pyplot as plt

if 'sentiment' in df.columns:
    class_counts = df['sentiment'].value_counts().sort_index()

    print("Sentiment Class Distribution:")
    for sentiment_class, count in class_counts.items():
        print(f"  Class {sentiment_class}: {count} samples")
    
    print(f"\nTotal samples: {class_counts.sum()}\n")

    plt.figure(figsize=(6, 4))
    class_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title("Distribution of Sentiment Classes")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
else:
    print("Column 'sentiment' not found in the dataset.")
```

## 6. Text Length (Word) Analysis

Here, we compute the word count for each message, then show average, median, min, and max. Finally, we display a histogram of word counts.

```python
if 'text' in df.columns:
    df['word_count'] = df['text'].astype(str).apply(lambda x: len(x.split()))
    
    avg_length = df['word_count'].mean()
    median_length = df['word_count'].median()
    max_length = df['word_count'].max()
    min_length = df['word_count'].min()
    
    print("Text Length Analysis (in words):")
    print(f"Average Length: {avg_length:.2f}")
    print(f"Median Length: {median_length}")
    print(f"Maximum Length: {max_length}")
    print(f"Minimum Length: {min_length}")
    
    plt.figure(figsize=(8, 5))
    plt.hist(df['word_count'], bins=30, color='lightgreen', edgecolor='black')
    plt.title("Histogram of Text Word Counts")
    plt.xlabel("Word Count")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'text' not found in the dataset.")
```

## 7. Word Distribution Visualization Using Word Cloud

We generate a word cloud from all messages to visualize the most common words.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Word Cloud for Each Class
if 'text' in df.columns and 'sentiment' in df.columns:
    for class_value in df['sentiment'].unique():
        # Filter the DataFrame for the current sentiment class
        subset_df = df[df['sentiment'] == class_value]
        
        # Combine all text entries into a single string
        text_combined = " ".join(subset_df['text'].dropna().astype(str).tolist())
        
        # Generate the word cloud
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_combined)
        
        # Display the word cloud
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(f"Word Cloud for sentiment = {class_value}")
        plt.axis("off")
        plt.show()
else:
    print("Either 'text' or 'sentiment' column not found in the dataset.")
```
