# Balanced Dataset Exploration

This notebook explores the balanced CSV file generated from the preprocessed code review dataset. 

It performs the following steps:
1. Loads the CSV file.
2. Displays the first 10 rows.
3. Samples 3 rows per class.
4. Analyzes the distribution of classes.
5. Analyzes text length with summary statistics and a histogram.
6. Generates a word cloud for each class.

## 1. Importing Required Libraries

We import the necessary libraries including `pandas`, `numpy`, `matplotlib`, and `wordcloud`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Enable inline plotting for Jupyter notebooks
%matplotlib inline
```

## 2. Loading the Balanced CSV Dataset

We load the CSV file generated from the preprocessing script. Update the path below if necessary.

```python
csv_file_path = "/Users/tayebekavousi/Desktop/github_sa/datasets/preprocessed/combined_dataset-balanced.csv"
df = pd.read_csv(csv_file_path)
print("Dataset loaded successfully!")
print("Dataset shape:", df.shape)
```

## 3. Displaying the First 10 Rows

The following cell prints the first 10 rows to give an initial view of the dataset.

```python
print("Head of the dataset (10 rows):")
display(df.head(10))
```

## 4. Sampling 3 Rows from Each Class

We sample 3 rows per class (based on the `is_toxic` label) to inspect representative examples.

```python
if 'sentiment' in df.columns:
    try:
        sample_df = df.groupby('sentiment').apply(lambda x: x.sample(n=3, random_state=42)).reset_index(drop=True)
        print("Sample (3 rows per class):")
        display(sample_df)
    except ValueError as e:
        print("Sampling error:", e)
else:
    print("Column 'sentiment' not found in the dataset.")
```

## 5. Distribution Analysis of Classes

We visualize the distribution of `is_toxic` classes using a bar chart.

```python
if 'sentiment' in df.columns:
    class_counts = df['sentiment'].value_counts()
    
    plt.figure(figsize=(6, 4))
    class_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title("Distribution of sentiment Classes")
    plt.xlabel("sentiment")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
else:
    print("Column 'sentiment' not found in the dataset.")
```

## 6. Text Length Analysis

We compute the word count for each message, print summary statistics (average, median, maximum, minimum), and display a histogram.

```python
if 'message' in df.columns:
    df['word_count'] = df['message'].astype(str).apply(lambda x: len(x.split()))
    
    avg_length = df['word_count'].mean()
    median_length = df['word_count'].median()
    max_length = df['word_count'].max()
    min_length = df['word_count'].min()
    
    print("Text Length Analysis (in words):")
    print(f"Average Length: {avg_length:.2f}")
    print(f"Median Length: {median_length}")
    print(f"Maximum Length: {max_length}")
    print(f"Minimum Length: {min_length}")
    
    plt.figure(figsize=(8, 5))
    plt.hist(df['word_count'], bins=30, color='lightgreen', edgecolor='black')
    plt.title("Histogram of Message Word Counts")
    plt.xlabel("Word Count")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Column 'message' not found in the dataset.")
```

## 7. Word Cloud for Each Class

We generate a word cloud for each `is_toxic` class to visualize the most common words.

```python
if 'message' in df.columns and 'sentiment' in df.columns:
    for class_value in df['sentiment'].unique():
        subset_df = df[df['sentiment'] == class_value]
        text_combined = " ".join(subset_df['message'].dropna().astype(str).tolist())
        
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_combined)
        
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(f"Word Cloud for sentiment = {class_value}")
        plt.axis("off")
        plt.show()
else:
    print("Either 'message' or 'sentiment' column not found in the dataset.")
```
